### Impute Missing Values
filling the missing values with Mschine Learning Algorithm

```python
from sklearn.experimental import enable_iterative_imputer

from sklearn.impute import IterativeImputer

from sklearn.ensemble import RandomForestRegressor
```

```python
df1=df.copy()
```

```python
# Separate numerical and categorical columns
numerical_cols=df1.select_dtypes(include=['number']).columns
categorical_cols=df1.select_dtypes(exclude=['number']).columns
```

```python
imputer_num=IterativeImputer(estimator=RandomForestRegressor(),random_state=0)
df1[numerical_cols]=imputer_num.fit_transform(df1[numerical_cols])

for col in categorical_cols:
    df1[col]=df1[col].fillna(df1[col].mode()[0])

print("\nMissing Values After Imputation:\n",df1.isnull().sum())
```

# Conclusion and Analysis of Titanic Dataset EDA
The exploratory data analysis (EDA) of the Titanic dataset revealed distinct patterns in survival outcomes influenced by demographic, socioeconomic, and situational factors:
### 1. Demographics
Gender played a crucial role: females had a significantly higher chance of survival than males.

Children (≤12 years) were prioritized during evacuation, showing the highest survival rate.

Older adults (60+ years) faced lower survival rates.

### 2. Socioeconomic Factors
Passenger class (Pclass) strongly influenced survival. First-class passengers had the highest survival rate due to better cabin locations and quicker access to lifeboats.

Fare was positively correlated with survival, highlighting that wealthier passengers had better chances of rescue.

### 3. Family and Companions
Passengers with small family sizes (1–3) had better survival outcomes.

Those traveling alone or with larger families (5+) were at a disadvantage, possibly due to difficulties in communication and coordination during evacuation.

 ### 4. Embarkation Port
Passengers who boarded from Cherbourg (C) had better survival chances, suggesting they were more likely to be in higher classes or better-located cabins.

Southampton (S) and Queenstown (Q) embarkees had comparatively lower survival rates.

### 5. Correlation Insights
Fare and Survived showed a strong positive correlation, reinforcing the influence of financial status.

There was a mild correlation between Pclass and Fare, indicating overlap between wealth and class.
